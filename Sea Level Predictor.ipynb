### Instructions

• Create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

• Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

• Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

• The x label should be Year, the y label should be Sea Level (inches), and the title should be Rise in Sea Level.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

df=pd.read_csv(r"your_path\epa-sea-level.csv")

# Create scatter plot
df.plot(kind='scatter', x="Year", y="CSIRO Adjusted Sea Level", title="Sea Level Scatter Plot")
```

```python
# Create first line of best fit
reg= linregress(df["Year"], df["CSIRO Adjusted Sea Level"]) #Linnear regression
x_pred=np.arange(df["Year"].min(), 2051)
y_pred=reg.intercept + reg.slope * x_pred
plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], s=15, label="original data")
plt.plot(x_pred, y_pred, color="red", label="predicted")
plt.xlabel("Year")
plt.ylabel("Sea Level")
plt.legend()
plt.title("Rise in Sea Level")
plt.show()
```

```python
#Second linear regression
mask=df[df["Year"]>=2000]
reg2= linregress(mask["Year"], mask["CSIRO Adjusted Sea Level"]) #Second Linear regression

x_pred_2=np.arange(mask["Year"].min(), 2051)
y_pred_2=reg2.intercept + reg2.slope * x_pred_2

plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], s=15, label="original data")
plt.plot(x_pred_2, y_pred_2, color="green", label="predicted 2000")
plt.legend()
plt.title("Rise in Sea Level")
plt.xlabel("Year")
plt.ylabel("Sea Level")
plt.show()
```
